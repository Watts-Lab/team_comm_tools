```python
import pandas as pd
import numpy as np
from utils.preprocess import *
```

```python
chat_data = pd.read_csv('./data/raw_data/csopII_conversations_withblanks.csv')
chat_data
```

```python
# go through all preprocessing steps
chat_data = preprocess_conversation_columns(chat_data)
assert_key_columns_present(chat_data)

# create new column that retains punctuation
chat_data["message_lower_with_punc"] = chat_data["message"].astype(str).apply(preprocess_text_lowercase_but_retain_punctuation)

# Preprocessing the text in `col` and then overwriting the column `col`.
# TODO: We should probably use classes to abstract preprocessing module as well?
chat_data["message"] = chat_data["message"].astype(str).apply(preprocess_text)
chat_data
```

```python
def get_turn_id(df) :
    df["speaker_nickname_x"] = df["speaker_nickname"].shift()
    return (df["speaker_nickname"] != df["speaker_nickname_x"]).cumsum()
    
turn_id_per_conv = chat_data.groupby(['conversation_num'], sort=False).apply(lambda df : get_turn_id(df))
turn_id_per_conv = turn_id_per_conv.to_frame().reset_index().rename(columns={0:'turn_id'})
turn_id_per_conv
```

```python
chat_data = pd.concat([chat_data, turn_id_per_conv["turn_id"]], axis=1)
```

```python
def compress(turn_df):
    result = turn_df.iloc[0]
    if (len(turn_df) > 1):
        result['message'] = turn_df['message'].str.cat(sep=' ')
        result['message_lower_with_punc'] = turn_df['message_lower_with_punc'].str.cat(sep=' ')
    return result

chat_data = chat_data.groupby('conversation_num', sort=False).apply(lambda df : df.groupby('turn_id', as_index=False).apply(compress)).reset_index(drop=True)
```
