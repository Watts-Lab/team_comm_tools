```python
import pandas as pd
import numpy as np
```

```python
user_df = pd.read_csv('./output/csopII_output_user_level.csv')
user_df
```

```python
speaker_lists = user_df.groupby(["conversation_num"]).apply(lambda df : np.asarray(df["speaker_nickname"]))

user_df = pd.merge(
    left=user_df,
    right=speaker_lists.rename("speaker_list"),
    on=['conversation_num'],
    how="inner"
)
user_df
```

```python
def remove_active_user(df):
    counter = 0
    for index, row in df.iterrows():
        user = row['speaker_nickname']
        speaker_list = row['speaker_list']
        i = np.argwhere(speaker_list==user)
        df.at[index,'speaker_list'] = np.delete(row['speaker_list'],i)
        df.iloc
    return df

user_df = user_df.groupby(["conversation_num"]).apply(lambda df : remove_active_user(df)).reset_index(drop=True)
user_df
```
