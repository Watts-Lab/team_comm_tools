```python
# sort the lexicons
certainty = pd.read_csv("./features/lexicons/certainty.txt").sort_values(["NumWords", "NumCharacters"], ascending=False)
certainty
```

```python
master_regex = certainty["Word"].str.cat(sep='|')
test = "i think that i think i think "
print(re.findall(master_regex, test))
```

```python
def get_certainty(chat_certainty) -> int:
    # iterate through the lexicons...
    certainty_score = 0
    
    # sub every instance with the associated normative certainty score 
    for index, row in certainty.iterrows():
        
        certainty_pattern = row["Word"]
        num_matches = len(re.findall(certainty_pattern, chat_certainty))
        if (num_matches > 0):
            print(f" this is the pattern: {certainty_pattern}")
        chat_certainty = re.sub(certainty_pattern, '', chat_certainty)
        certainty_score += (row["Certainty"] * num_matches)
       
    if (certainty_score > 0):
        print(certainty_score)
    return certainty_score
```

```python
def get_certainty(chat_certainty) -> int:
    # iterate through the lexicons...
    certainty_score = 0
    matches = re.findall(master_regex, chat_certainty)

    for match in matches:
        certainty_score += certainty.loc[certainty['Word'] == match]["Certainty"].iloc[0]

    if (len(matches) == 0):
        return certainty_score
    return (certainty_score / len(matches))
```

```python
df["certainty"] = df["message"].apply(get_certainty)
df
```
