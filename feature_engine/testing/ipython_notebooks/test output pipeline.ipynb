```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
jury_raw = pd.read_csv('../../data/raw_data/jury_conversations_with_outcome_var.csv')
csop_raw = pd.read_csv('../../data/raw_data/csop_conversations_withblanks.csv')
```

```python
jury_features = pd.read_csv('../../output/jury_output_conversation_level.csv')
csop_features = pd.read_csv('../../output/csop_withblanks_output_conversation_level.csv')
```

```python
juries = pd.merge(
    left=jury_features,
    right = jury_raw[["batch_num", "round_num", "majority_pct", "num_flipped", "flipped_pct", "num_votes"]],
    on=['batch_num', 'round_num'],
    how="left"
).drop_duplicates()
```

```python
juries.head()
```

```python
csop = pd.merge(
    left=csop_features,
    right = csop_raw[["batch_num", "round_num", "type", "social_perceptiveness", "skill", "normalized_score", "zscore_score", "zscore_round_duration", "zscore_efficiency"]],
    on=['batch_num', 'round_num'],
    how="left"
).drop_duplicates()
```

```python
csop.head()
```

```python
# export some csv's to play with modeling
juries.to_csv('juries_conv_for_modeling.csv')
csop.to_csv('csop_conv_for_modeling_withblanks.csv')
```
